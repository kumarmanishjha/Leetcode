Given an array of intervals where 

intervals[i] = [start_i, end_i], 

merge all overlapping intervals, and return an array of the non-overlapping intervals that cover all the intervals in the input.

```python
from typing import List
```

```python
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        merged_list = []
        min_start = intervals[0][0]
        max_end = intervals[0][-1]
        for i in range(len(intervals) -1):
            min_start = min(min_start, intervals[i+1][0])
            if(max_end >= intervals[i+1][0]):
                max_end = max(max_end, intervals[i+1][-1])
            else:
                merged_list.append([min_start, max_end])
                if(i+1 < len(intervals)):
                    min_start = intervals[i+1][0]
                    max_end =intervals[i+1][-1]
        
        merged_list.append([min_start, max_end])
        
        return merged_list
            
```

```python
S= Solution()
```

```python
intervals = [[1,3], [2,6],[8,10], [15,18]]
```

```python
S.merge(intervals)
```
