```python
from typing import List
```

```python
class Solution:
    def getSkyline(self, buildings: List[List[int]]) -> List[List[int]]:
        skyline = []
        skyline.append([buildings[0][0], buildings[0][-1]])
        
        for i in range(len(buildings) - 1):
            if(buildings[i][1] < buildings[i+1][0]):
                skyline.append([buildings[i][1], 0])
                skyline.append([buildings[i+1][0], buildings[i+1][-1]])
                
            elif(buildings[i][1] > buildings[i+1][0] and buildings[i+1][-1] > buildings[i][-1]):
                skyline.append([buildings[i+1][0], buildings[i+1][-1]])
                
            elif(buildings[i][1] > buildings[i+1][0] and buildings[i+1][-1] < buildings[i][-1]):
                skyline.append([buildings[i][1], buildings[i+1][-1]])
                
            else:
                continue

        
        skyline.append([buildings[-1][1], 0])
        
        return skyline
```

```python
S = Solution()
```

```python
S.getSkyline(buildings=[[0,2,3],[2,5,3],[3,7,4]])
```
