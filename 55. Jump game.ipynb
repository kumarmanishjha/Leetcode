```python
from typing import List
```

```python
class Solution:
    def canJump(self, nums: List[int]) -> bool:
        
        if (0 in nums):
            i = len(nums) - 1
            stuck_at = i
            while(i>=0):
                curr = nums[i]
                if (i + curr >= stuck_at):
                    stuck_at = i
                i = i-1

            if(stuck_at > 0):
                return False
        
        return True
        

        
```

```python
nums = [2,0,8,1,1,1,0,2,0,5,3,2,1,0,4]
#nums = [2,3,1,1,4]
#nums = [1,1,0,5,3,2,1,0,4]
```

```python
S = Solution()
```

```python
S.canJump(nums)
```
